```python
import csv
import pandas as pd
import numpy as np

############################################
## ADD THE COUNTRY THAT U WANNA TO FILTER ##
############################################

aim_country = " Peru"
```

```python
### UPLOADING THE RESPONSE ###
csvfile = "response.csv"
```

```python
### READING THE RESPONSE AS A DATAFRAME ###
df = pd.read_csv(csvfile)
```

```python
### SHOWING THE COLUMN THAT HAS THE ADDRESS ###
df['data__formattedAddress']
```

```python
### SPLITING THE ADDRESS TO TAKE THE COUNTRY ###
df['AB_split'] = df['data__formattedAddress'].str.split(',')
countries = []
```

```python
### PUTING THE COUNTRIES IN A ARRAY ###
for i in range(0, total_rows):
         countries.append(df['AB_split'][i][-1])
#         df['country'] = pd.Series((df['AB_split'][i][-1]), index=i)
         print(df['AB_split'][i][-1])
```

```python
### TRANSFORMING THE ARRAY OF COUNTRIES IN A SERIES ###
p = pd.Series(countries, index=range(0,total_rows))
```

```python
p
```

```python
### CREATING A NEW COLUMN, OF COUNTRIES, IN THE DATAFRAME ###
df['countries'] = p
```

```python
df['countries']
```

```python
### CREATING A NEW DATAFRAME THAT HAVE ONLY THE DATA WITH THE SEARCH COUNTRY ###
df_C = df[df['countries'] == aim_country]
```

```python
### EXPORTING AS A .csv FILE ### 
export_csv = df_C.to_csv (r'export_dataframe_PERU_2.csv', index = True, header=True) #Don't forget to add '.csv' at the end of the path
```
